```python
import matplotlib.pyplot as plt
```

```python
# Data for each crime offense
offenses = [
    "Aggravated Assault", "Burglary", "Larceny",
    "Motor Vehicle Theft", "Murder", "Rape (revised definition)", "Robbery"
]
rates_2023 = [264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5]

# Customize bar colors
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

# Create the bar chart
plt.figure(figsize=(10, 6))
bars = plt.bar(offenses, rates_2023, color=colors)

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Label the axes and add a title
plt.xlabel("Type of Offense")
plt.ylabel("Offenses per 100,000 People (2023)")
plt.title("Crime Rates per 100,000 People in 2023")

# Rotate x-axis labels for better visibility
plt.xticks(rotation=45, ha='right')

# Display the plot
plt.tight_layout()
plt.show()
```

```python
# Data for each crime offense
offenses = [
    "Aggravated Assault", "Burglary", "Larceny",
    "Motor Vehicle Theft", "Murder", "Rape (revised definition)", "Robbery"
]
# Change in offense rates from 2022 to 2023 (absolute change per 100,000)
change_rates = [-3.3, -8.1, -4.9, +12.0, -12.0, -9.8, -0.8]

# Unique colors for each offense
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2']

plt.figure(figsize=(10, 6))
bars = plt.barh(offenses, change_rates, color=colors)

# Add gridlines for better readability
plt.grid(axis='x', linestyle='--', alpha=0.7)

# Label the axes and add a title
plt.xlabel("Change in Offenses per 100,000 People (2022 to 2023)")
plt.title("Change in Crime Rates (2022 vs. 2023)")

plt.tight_layout()
plt.show()
```

```python
import numpy as np
# Data for each crime offense
offenses = [
    "Aggravated Assault", "Burglary", "Larceny",
    "Motor Vehicle Theft", "Murder", "Rape (revised definition)", "Robbery"
]
# Offense rates per 100,000 people for 2023
rates_2023 = np.array([264.1, 250.7, 1347.2, 318.7, 5.7, 38.0, 66.5])
# Change in offense rates from 2022 to 2023 (2022 to 2023 = 2023 rate - 2022 rate)
change = np.array([-3.3, -8.1, -4.9, +12.0, -12.0, -9.8, -0.8])
# Calculate 2022 rates
rates_2022 = rates_2023 - change  # Rearranged: 2022 = 2023 - (change)

# X locations for the groups
x = np.arange(len(offenses))
width = 0.35  # Width of the bars

plt.figure(figsize=(12, 7))

# Plot bars for 2022 and 2023 side-by-side
bars1 = plt.bar(x - width/2, rates_2022, width, label='2022', color='#1f77b4')
bars2 = plt.bar(x + width/2, rates_2023, width, label='2023', color='#ff7f0e')

# Add gridlines for better readability
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Label the axes and add a title
plt.xlabel("Type of Offense")
plt.ylabel("Offenses per 100,000 People")
plt.title("Comparison of Crime Rates for 2022 and 2023")
plt.xticks(x, offenses, rotation=45, ha='right')

# Add a legend to indicate which bars represent which year
plt.legend()

plt.tight_layout()
plt.show()
```
