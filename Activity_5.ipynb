```python
# State Data Interactive Query Program

import pandas as pd

#  Load the CSV File
data = pd.read_csv('State_Data.csv')

#  Display Available Columns and Options
print("Available Data Columns:")
print(data.columns.tolist())

#  Simple Interactive Query System
def query_data():
    while True:
        print("\nOptions: \n1. Lookup by State \n2. Lookup by Region \n3. Exit")
        choice = input("Enter your choice (1/2/3): ")

        if choice == '1':
            state_name = input("Enter state name: ").strip()
            result = data[data['State'].str.lower() == state_name.lower()]
            if result.empty:
                print("State not found.")
            else:
                print(result.to_string(index=False))

        elif choice == '2':
            region_name = input("Enter region name (e.g., Southeast, Far_West): ").strip()
            result = data[data['Region'].str.lower() == region_name.lower()]
            if result.empty:
                print("Region not found.")
            else:
                print(result[['State', 'Population (millions)', 'GDP (billions)']].to_string(index=False))

        elif choice == '3':
            print("Exiting the program. Goodbye!")
            break
        else:
            print("Invalid option. Try again.")

# Step 4: Start the Query Loop
query_data()
```
