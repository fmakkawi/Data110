```python
# Author: Farah Makkawi

# Project 2 - Calories Estimations

# Calorie data for food and add-ons
food_items = {
    "burger": 500,
    "chicken salad": 350,
    "fish tacos": 400,
    "veggie wrap": 300
}

add_ons = {
    "chips": 150,
    "fries": 300,
    "drink": 150,
    "dressing": 100,
    "dessert": 350
}

recommended_calories = {
    "adult": 2000,
    "child": 1400
}

# Main function to calculate meal calories
def calorie_estimator():
    print("Welcome to the Calorie Estimator!\n")

    while True:
        print("Are you ordering for an adult or child?")
        person_type = input("Enter 'adult' or 'child': ").strip().lower()

        if person_type not in recommended_calories:
            print("Invalid entry. Please try again.\n")
            continue

        total_calories = 0
        order_list = []

        # Get food choice
        print("\nAvailable food items:")
        for item in food_items:
            print(f"- {item.title()}")

        while True:
            food = input("\nEnter a food item (or type 'done' to finish): ").strip().lower()
            if food == 'done':
                break
            if food in food_items:
                total_calories += food_items[food]
                order_list.append(food.title())
            else:
                print("Item not recognized, try again.")

        if not order_list:
            print("You must select at least one food item.\n")
            continue

        # Get optional add-ons
        print("\nOptional Add-ons:")
        for addon in add_ons:
            print(f"- {addon.title()}")

        while True:
            addon = input("\nEnter an add-on (or type 'none' to skip): ").strip().lower()
            if addon == 'none':
                break
            if addon in add_ons:
                total_calories += add_ons[addon]
                order_list.append(addon.title())
            else:
                print("Add-on not recognized, try again.")

        # Output the result
        print("\nYour Order Summary:")
        for item in order_list:
            print(f"- {item}")
        print(f"Total Calories: {total_calories} cal")

        # Recommendation
        if total_calories > recommended_calories[person_type]:
            print("‚ö†Ô∏è Your meal exceeds the recommended daily calories.")
            print("üí° Suggestion: Consider reducing portions or choosing healthier options.\n")
        else:
            print("‚úÖ Your meal is within a healthy calorie range.\n")

        # Repeat?
        again = input("Would you like to estimate calories for another meal? (yes/no): ").strip().lower()
        if again != 'yes':
            print("Thank you for using the Calorie Estimator!")
            break
calorie_estimator()
```
