```python
from google.colab import files

# Upload boy names file
print("Upload the boy names file:")
uploaded_boy = files.upload()

# Upload girl names file
print("Upload the girl names file:")
uploaded_girl = files.upload()
```

```python
# Read boy names
with open(list(uploaded_boy.keys())[0], 'r') as bf:
    boy_names = [line.strip() for line in bf.readlines()]

# Read girl names
with open(list(uploaded_girl.keys())[0], 'r') as gf:
    girl_names = [line.strip() for line in gf.readlines()]

# Combine into master list
master_list = []
for name in boy_names:
    master_list.append({'name': name, 'gender': 'Boy'})

for name in girl_names:
    master_list.append({'name': name, 'gender': 'Girl'})

# Check sample
master_list[:5]
```

```python
# Search by gender
def search_by_gender(gender):
    results = [entry for entry in master_list if entry['gender'].lower() == gender.lower()]
    return results

# Exact match search
def exact_match(name):
    results = [entry for entry in master_list if entry['name'].lower() == name.lower()]
    return results

# Search names starting with specific letters
def starts_with(prefix):
    results = [entry for entry in master_list if entry['name'].lower().startswith(prefix.lower())]
    return results

# Extra feature: Search names containing letters
def contains_substring(substring):
    results = [entry for entry in master_list if substring.lower() in entry['name'].lower()]
    return results
```

Menu System for User Interaction

```python
while True:
    print("\n--- Baby Names Search ---")
    print("1. Search by Gender (Boy/Girl)")
    print("2. Exact Name Match")
    print("3. Names Starting With")
    print("4. Names Containing Letters")
    print("5. Exit")

    choice = input("Choose an option (1-5): ")

    if choice == '1':
        gender = input("Enter gender (Boy/Girl): ")
        results = search_by_gender(gender)
        print(f"Found {len(results)} names:\n", results[:10])

    elif choice == '2':
        name = input("Enter the exact name: ")
        results = exact_match(name)
        print(f"Found {len(results)} matches:\n", results)

    elif choice == '3':
        prefix = input("Enter the starting letters: ")
        results = starts_with(prefix)
        print(f"Found {len(results)} names:\n", results[:10])

    elif choice == '4':
        substring = input("Enter the letters to search within names: ")
        results = contains_substring(substring)
        print(f"Found {len(results)} names:\n", results[:10])

    elif choice == '5':
        print("Goodbye!")
        break

    else:
        print("Invalid option, try again.")
```
