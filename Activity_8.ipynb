```python
# Author: Farah Makkawi

import pandas as pd
import matplotlib.pyplot as plt

# Sample dataset
data = {
    'Student': ['A', 'B', 'C', 'D', 'E'],
    'Hours_Studied': [5, 8, 3, 6, 9],
    'Final_Grade': [75, 90, 60, 80, 95]
}

df = pd.DataFrame(data)

# Chart 1: Scatter plot
plt.figure(figsize=(6, 4))
plt.scatter(df['Hours_Studied'], df['Final_Grade'], color='blue')
plt.title('Hours Studied vs Final Grade')
plt.xlabel('Hours Studied')
plt.ylabel('Final Grade')
plt.grid(True)
plt.show()

# Chart 2: Bar plot
plt.figure(figsize=(6, 4))
plt.bar(df['Student'], df['Final_Grade'], color='green')
plt.title('Final Grades of Students')
plt.xlabel('Student')
plt.ylabel('Final Grade')
plt.show()
```
