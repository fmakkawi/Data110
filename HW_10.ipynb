```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("carsData.csv", header=None)

df.head()
```

Clean and Rename the Columns

```python
# Temporarily rename the columns
df.columns = ["col1", "col2", "col3", "col4", "col5", "col6", "col7", "col8"]

# Remove the first row which contains the real column names
df = df[1:].copy()

# Assign real column names
df.columns = ["mpg", "engine", "horse", "weight", "acceleration", "year", "origin", "cylinder"]

# Convert all columns to numeric values
df = df.apply(pd.to_numeric, errors="coerce")

# Display cleaned data
df.head()
```

Analyze the Data with Pandas

```python
# Average MPG by cylinder count
avg_mpg_by_cyl = df.groupby("cylinder")["mpg"].mean()
print("Average MPG by Cylinder:\n", avg_mpg_by_cyl)

# Average horsepower by engine size
avg_hp_by_engine = df.groupby("engine")["horse"].mean()
print("\nAverage Horsepower by Engine:\n", avg_hp_by_engine.head())

# Average weight by year
avg_weight_by_year = df.groupby("year")["weight"].mean()
print("\nAverage Weight by Year:\n", avg_weight_by_year.head())
```

Create Chart 1 – MPG by Cylinders

```python
plt.figure(figsize=(6, 4))
avg_mpg_by_cyl.plot(kind='bar', color='skyblue')
plt.title('Average MPG by Number of Cylinders')
plt.xlabel('Cylinders')
plt.ylabel('Average MPG')
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

Create Chart 2 – Weight by Year

```python
plt.figure(figsize=(6, 4))
avg_weight_by_year.plot(kind='line', marker='o', color='green')
plt.title('Average Car Weight by Year')
plt.xlabel('Year')
plt.ylabel('Average Weight (lbs)')
plt.grid(True)
plt.tight_layout()
plt.show()
```
