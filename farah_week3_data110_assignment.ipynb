```python
import matplotlib.pyplot as plt
import numpy as np

# Print available styles
print(plt.style.available)

# Set the style to 'seaborn-v0_8-whitegrid'
plt.style.use('seaborn-v0_8-whitegrid')

# Choose another style from the available list (for later use, let's choose 'ggplot')
chosen_style = 'ggplot'

# Define the quadratic function
def f(x):
    return x**2 - 2*x + 1

# Generate x values from -2 to 4
x = np.linspace(-2, 4, 100)

# Generate y values using the function f(x)
y = f(x)

# Create the plot
plt.plot(x, y)

# Add labels and a title
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of the quadratic function f(x) = x^2 - 2x + 1')

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
    return x**3

def g(x):
    return np.exp(x)

def h(x):
    return np.log(x)

# Generate x values
x1 = np.linspace(-2, 2, 100)  # For f(x) = x^3
x2 = np.linspace(0, 2, 100)   # For g(x) = e^x and h(x) = log(x), we avoid x <= 0 for log

# Calculate y values for each function
y_f = f(x1)
y_g = g(x2)
y_h = h(x2)

# Create the plot
plt.plot(x1, y_f, label='$f(x) = x^3$', color='b')  # Blue line for f(x)
plt.plot(x2, y_g, label='$g(x) = e^x$', color='r')  # Red line for g(x)
plt.plot(x2, y_h, label='$h(x) = \log(x)$', color='g')  # Green line for h(x)

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f(x), g(x), and h(x) on the same graph')
plt.legend()

# Display the plot
plt.show()
```

```python


# Define the functions
def f(x):
    return x**2

def g(x):
    return x**3

def h(x):
    return x**4

def i(x):
    return x**5

# Generate x values
x = np.linspace(-2, 2, 100)

# Calculate y values for each function
y_f = f(x)
y_g = g(x)
y_h = h(x)
y_i = i(x)

# Create the plot
plt.plot(x, y_f, label='$f(x) = x^2$', color='red', linestyle='-', linewidth=2)    # Named color: red, solid line
plt.plot(x, y_g, label='$g(x) = x^3$', color='#FF5733', linestyle='--', linewidth=2)  # Hex color, dashed line
plt.plot(x, y_h, label='$h(x) = x^4$', color=(0.1, 0.2, 0.5), linestyle='-.', linewidth=2)  # RGB tuple, dash-dot line
plt.plot(x, y_i, label='$i(x) = x^5$', color='g', linestyle=':', linewidth=2)  # Shorthand color code, dotted line

# Add labels, title, and legend
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot of f(x), g(x), h(x), and i(x) with Different Colors and Line Styles')
plt.legend()

# Display the plot
plt.show()
```

```python

# Define the function f(x) = 1 / (x^2 + 1)
def f(x):
    return 1 / (x**2 + 1)

# Generate x values over the range [-10, 10]
x = np.linspace(-10, 10, 400)

# Calculate y values using the function f(x)
y = f(x)

# Plot 1: Function with custom axes limits
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = \\frac{1}{x^2 + 1}$', color='b')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = \\frac{1}{x^2 + 1}$')
plt.xlim(-10, 10)  # Custom x-axis limits
plt.ylim(0, 1)    # Custom y-axis limits
plt.legend()

# Display the plot
plt.show()

# Plot 2: Function with reversed x and y axes
plt.figure(figsize=(10, 6))
plt.plot(y, x, label='$f(x) = \\frac{1}{x^2 + 1}$ (reversed axes)', color='r')
plt.xlabel('f(x)')
plt.ylabel('x')
plt.title('Reversed Axes: $f(x) = \\frac{1}{x^2 + 1}$')
plt.xlim(0, 1)  # Custom x-axis limits for reversed axes plot
plt.ylim(-10, 10)  # Custom y-axis limits for reversed axes plot
plt.legend()

# Display the plot
plt.show()

# Plot 3: Function with 'tight' and 'equal' axis settings
# 'tight' adjusts axis to fit the data closely, 'equal' ensures equal scaling of both axes
plt.figure(figsize=(10, 6))
plt.plot(x, y, label='$f(x) = \\frac{1}{x^2 + 1}$', color='g')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Plot of $f(x) = \\frac{1}{x^2 + 1}$ with Tight and Equal Axis')

# Applying 'tight' and 'equal' axes
plt.axis('tight')  # Automatically adjust axes to fit the data
plt.axis('equal')  # Make the x and y axes have equal scaling

plt.legend()

# Display the plot
plt.show()
```

```python


# Define the function f(x) = x * cos(x)
def f(x):
    return x * np.cos(x)

# Generate x values over the range [-2*pi, 2*pi]
x = np.linspace(-2 * np.pi, 2 * np.pi, 400)  # 400 points for smooth curve

# Calculate y values using the function f(x)
y = f(x)

# Generate x values for scatter points at every π/4 interval
x_scatter = np.arange(-2 * np.pi, 2 * np.pi + np.pi / 4, np.pi / 4)  # x values for scatter
y_scatter = f(x_scatter)  # Calculate corresponding y values for scatter points

# Create the plot
plt.plot(x, y, label='$f(x) = x \\cos(x)$', color='b', linestyle='-', linewidth=2)  # Line plot
plt.scatter(x_scatter, y_scatter, color='r', label='Scatter Points', zorder=5, s=50)  # Scatter plot

# Add title and labels
plt.title('Combined Plot of $f(x) = x \\cos(x)$ with Scatter Points at $\\frac{\\pi}{4}$ intervals')
plt.xlabel('x')
plt.ylabel('f(x)')

# Add legend
plt.legend()

# Show the plot
plt.show()
```

```python


# Choose alternative style (e.g., 'ggplot')
plt.style.use('ggplot')

# Define the functions
def f(x):
    return np.tan(x)

def g(x):
    return x * np.exp(-x**2)

def h(x):
    # Handling x = 0 for sin(x) * x
    return np.sin(x) * x

def i(x):
    return np.arctan(x)

# Generate x values
x = np.linspace(-5, 5, 400)
x_fine = np.linspace(-5, 5, 1000)  # For smooth tan curve
x_scatter = np.arange(-5, 5, 1)  # Scatter points at integer x values

# Calculate y values using the functions
y_f = f(x_fine)
y_g = g(x)
y_h = h(x)
y_i = i(x)

# Create the plot
plt.plot(x_fine, y_f, label='$f(x) = \\tan(x)$', color='purple', linestyle='-', linewidth=2)  # tan(x) plot
plt.plot(x, y_g, label='$g(x) = x e^{-x^2}$', color='green', linestyle='--', linewidth=2)  # x * exp(-x^2) plot
plt.plot(x, y_h, label='$h(x) = \\sin(x) x$', color='blue', linestyle=':', linewidth=2)  # sin(x) * x plot
plt.plot(x, y_i, label='$i(x) = \\arctan(x)$', color='red', linestyle='-.', linewidth=2)  # arctan(x) plot

# Scatter points for y_g at specific x values (integer points)
plt.scatter(x_scatter, g(x_scatter), color='orange', label='Scatter points for $g(x)$', zorder=5, s=80)

# Customize limits to zoom into a region of interest (handle the tan function's asymptotes)
plt.xlim([-5, 5])
plt.ylim([-10, 10])

# Add title, labels, and legend
plt.title('Custom Plot with Multiple Functions', fontsize=14)
plt.xlabel('x', fontsize=12)
plt.ylabel('y', fontsize=12)
plt.legend()

# Display the plot
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Define the functions
def f(x):
    return x**3

def g(x):
    return np.exp(x)

def h(x):
    return np.log(x)

# Generate x values
x = np.linspace(-2, 4, 400)
x_positive = np.linspace(0.1, 4, 400)  # Avoid log(0)

# Calculate y values using the functions
y_f = f(x)
y_g = g(x)
y_h = h(x_positive)

# Create the plot for Default Style
plt.style.use('default')
plt.plot(x, y_f, label='$f(x) = x^3$', color='b', linestyle='-', linewidth=2)
plt.plot(x, y_g, label='$g(x) = e^x$', color='r', linestyle='--', linewidth=2)
plt.plot(x_positive, y_h, label='$h(x) = \log(x)$', color='g', linestyle=':', linewidth=2)

plt.title('Plot with Default Style')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Create the plot for Seaborn-V0_8-Whitegrid Style
plt.style.use('seaborn-v0_8-whitegrid')
plt.plot(x, y_f, label='$f(x) = x^3$', color='b', linestyle='-', linewidth=2)
plt.plot(x, y_g, label='$g(x) = e^x$', color='r', linestyle='--', linewidth=2)
plt.plot(x_positive, y_h, label='$h(x) = \log(x)$', color='g', linestyle=':', linewidth=2)

plt.title('Plot with Seaborn-V0_8-Whitegrid Style')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()

# Create the plot for the Chosen Alternative Style ('ggplot')
plt.style.use('ggplot')
plt.plot(x, y_f, label='$f(x) = x^3$', color='b', linestyle='-', linewidth=2)
plt.plot(x, y_g, label='$g(x) = e^x$', color='r', linestyle='--', linewidth=2)
plt.plot(x_positive, y_h, label='$h(x) = \log(x)$', color='g', linestyle=':', linewidth=2)

plt.title('Plot with ggplot Style')
plt.xlabel('x')
plt.ylabel('y')
plt.legend()
plt.show()
```
