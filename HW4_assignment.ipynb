```python
import pandas as pd

# Load file
df = pd.read_csv('Stocks.csv')
df.columns = df.columns.str.strip()
# Clean the columns
df['Purchase Price'] = df['Purchase Price'].str.replace('$', '').astype(float)
df['Current Price'] = df['Current Price'].str.replace('$', '').astype(float)
df['Shares'] = df['Shares'].astype(int)


df.dtypes

```

```python
def total_purchase(df):
    total_price = (df['Purchase Price'] * df['Shares']).sum()
    print("Total Purchase Price:", total_price)

total_purchase(df)


```

```python
def profit_margin(df):
    profit = ((df['Current Price'] - df['Purchase Price']) * df['Shares']).sum()
    print("Total Profit Margin:", profit)

profit_margin(df)
```

```python
# Count occurrences of a word
def foo(input_string, sub_string):
    count = input_string.count(sub_string)
    print(f"The word '{sub_string}' appeared {count} times.")

foo("Emma is smart. Emma learns fast. Emma is here.", "Emma")


# Count total words
def word_count(input_string):
    total = len(input_string.split())
    print("Total words:", total)

word_count("Emma is smart. Emma learns fast. Emma is here.")
```
