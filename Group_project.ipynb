```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
```

```python


df = pd.read_csv("Students Social Media Addiction.csv")
df.head()
```

```python
df_selected = df[['Avg_Daily_Usage_Hours', 'Sleep_Hours_Per_Night']].dropna()
df_selected.head()
```

```python


plt.figure(figsize=(8,6))
sns.scatterplot(x='Avg_Daily_Usage_Hours', y='Sleep_Hours_Per_Night', data=df_selected)
plt.title('Daily Social Media Use vs. Sleep Hours')
plt.xlabel('Average Daily Usage (Hours)')
plt.ylabel('Sleep Hours Per Night')
plt.show()
```

```python
from scipy.stats import pearsonr

corr_coef, p_value = pearsonr(df_selected['Avg_Daily_Usage_Hours'], df_selected['Sleep_Hours_Per_Night'])
print(f"Pearson correlation coefficient: {corr_coef:.2f}")
print(f"P-value: {p_value:.4f}")
```
