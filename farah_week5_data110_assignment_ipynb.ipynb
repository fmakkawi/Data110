```python
import seaborn as sns
import pandas as pd

# Load Titanic dataset
titanic = sns.load_dataset('titanic')

# View the first few rows of the dataset
titanic.head()
```

```python
dataset_summary = """
The Titanic dataset consists of 891 rows and 15 columns, containing information about passengers aboard the Titanic, such as demographic details, ticket class, fare, and survival status. The key variables include survived, indicating whether a passenger survived (1) or not (0); pclass, representing the ticket class (1st, 2nd, or 3rd); sex, denoting the passenger’s gender; and age, which contains some missing values. Other important columns are sibsp (number of siblings/spouses aboard), parch (number of parents/children aboard), fare (ticket price), and embarked (port of embarkation: C = Cherbourg, Q = Queenstown, S = Southampton). The dataset also includes categorical variables like deck, which has many missing values, and alone, indicating whether a passenger was traveling alone. Initial exploration reveals missing values in the age, deck, and embark_town columns, which may influence analysis. This dataset is useful for examining survival patterns based on different passenger attributes.
"""

print(dataset_summary)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Plot histogram and KDE for fare distribution
plt.figure(figsize=(10, 5))

sns.histplot(titanic['fare'], bins=30, kde=True, color='blue')

# Add labels and title
plt.xlabel('Fare')
plt.ylabel('Frequency')
plt.title('Distribution of Passenger Fares on the Titanic')

# Show the plot
plt.show()
```

### Question: What trends or outliers do you observe in the fare distribution? How might fare impact survival rates?

The distribution of fares is right-skewed, indicating that most passengers paid relatively low fares, with a few paying very high fares, creating noticeable outliers. The majority of passengers paid fares below $50, while a few outliers paid over $500, likely in first-class accommodations.

The fare may have a significant impact on survival rates, as higher fares were often associated with first-class passengers. First-class passengers had a higher chance of survival, as they had better access to lifeboats and were often prioritized during evacuation. On the other hand, third-class passengers, who generally paid lower fares, had a lower survival rate, likely due to being located in areas with less access to lifeboats and slower evacuation procedures.


```python

# Create a KDE plot comparing fare distributions for survivors and non-survivors
sns.kdeplot(data=titanic, x='fare', hue='survived', fill=True, common_norm=False, palette="crest")

# Adding labels and title
plt.title('Fare Distribution by Survival Status')
plt.xlabel('Fare')
plt.ylabel('Density')

# Show plot
plt.show()
```

### Question: Does fare seem to have influenced survival? What do the distributions suggest about socioeconomic status and survival?

From the KDE plot, we can observe that passengers who survived generally paid higher fares, especially compared to non-survivors. The distribution for survivors is shifted towards higher fare values, with a more prominent peak at higher fares, while the distribution for non-survivors is more spread out and includes more passengers who paid lower fares.

This suggests that fare may have influenced survival chances, as passengers in higher fare classes (such as first-class) were more likely to survive. This is consistent with historical evidence that passengers in first-class cabins had better access to lifeboats and were prioritized during the evacuation process. On the other hand, passengers who paid lower fares, especially those in third-class, had lower survival rates, which may reflect their location on the ship and the difficulties they faced in accessing lifeboats.


```python
# Plotting histograms of fare distributions across different passenger classes
plt.figure(figsize=(12, 6))

# Create the histograms
sns.histplot(data=titanic, x='fare', hue='pclass', kde=True, multiple="stack", palette="crest")
plt.title('Fare Distribution by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Count')
plt.show()

# Plotting KDE plots for each passenger class
plt.figure(figsize=(12, 6))
sns.kdeplot(data=titanic, x='fare', hue='pclass', fill=True, common_norm=False, palette="crest")
plt.title('Fare Distribution (KDE) by Passenger Class')
plt.xlabel('Fare')
plt.ylabel('Density')
plt.show()
```

```python
# Plotting survival rates by passenger class
plt.figure(figsize=(8, 6))
sns.countplot(data=titanic, x='pclass', hue='survived', palette='crest')
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Count')
plt.show()
```

### Question: What patterns do you observe in fare distribution and survival rates across different classes? How does this reflect the socioeconomic divisions on the Titanic?

From the histograms and KDE plots, we can observe significant differences in fare distributions across the three passenger classes (Pclass). Passengers in 1st class (Pclass=1) generally paid much higher fares compared to those in 2nd and 3rd class, with a concentration of fares around the $50-$100 range for first-class passengers and a wider spread of lower fares for second and third-class passengers. This reflects the socioeconomic divide on the Titanic, where first-class passengers were wealthier and likely had more access to amenities and safety measures.

In terms of survival rates, we observe a clear pattern where 1st class passengers had the highest survival rate, followed by 2nd class, and lastly, 3rd class passengers had the lowest survival rate. This is consistent with the notion that passengers in higher classes (first-class) were prioritized during the evacuation, having easier access to lifeboats and other safety measures. Conversely, third-class passengers, who were located in the lower parts of the ship, faced greater challenges in escaping during the disaster.

Overall, these patterns underscore the significant impact of socioeconomic status on survival chances during the Titanic disaster, with wealthier passengers having higher chances of survival, illustrating the stark class divisions that existed on the ship.


```python
# Plotting KDE plots for age distributions of survivors and non-survivors
plt.figure(figsize=(12, 6))
sns.kdeplot(data=titanic, x='age', hue='survived', fill=True, common_norm=False, palette="crest")
plt.title('Age Distribution of Survivors vs Non-Survivors')
plt.xlabel('Age')
plt.ylabel('Density')
plt.show()
```

### Question: What trends do you notice in the age distribution of survivors compared to those who did not survive? How might age have impacted survival rates?

The KDE plots reveal distinct differences in the age distributions between survivors and non-survivors. Survivors tend to have a higher concentration of younger passengers, particularly children and young adults. The peak of the age distribution for survivors is noticeably shifted toward younger age groups, with a significant portion of survivors being under 10 years old. In contrast, the age distribution for non-survivors is more spread out, with a larger proportion of adults in the middle-aged group.

This suggests that age may have played an important role in survival. Younger passengers, particularly children, were more likely to survive, which may reflect the priority given to children during the evacuation process. Additionally, it is possible that older or middle-aged passengers had a harder time navigating the ship or accessing lifeboats, which could explain the higher survival rates among younger passengers. These trends align with historical accounts of lifeboat priorities that favored women and children, as well as the physical challenges faced by older adults in a crisis situation.

In summary, age seems to have significantly influenced survival chances, with younger passengers having a higher likelihood of survival, while older passengers were more likely to have perished in the disaster.


```python
# Create the pyramid-style histogram
plt.figure(figsize=(12, 8))

# Separate the dataset into survivors and non-survivors
survivors = titanic[titanic['survived'] == 1]
non_survivors = titanic[titanic['survived'] == 0]

# Plot for male survivors and non-survivors
sns.histplot(survivors[survivors['sex'] == 'male']['age'], kde=False, bins=20, color='blue', label='Male Survivors', stat='density', alpha=0.6, multiple="stack")
sns.histplot(non_survivors[non_survivors['sex'] == 'male']['age'], kde=False, bins=20, color='red', label='Male Non-Survivors', stat='density', alpha=0.6, multiple="stack")

# Plot for female survivors and non-survivors
sns.histplot(survivors[survivors['sex'] == 'female']['age'], kde=False, bins=20, color='lightblue', label='Female Survivors', stat='density', alpha=0.6, multiple="stack")
sns.histplot(non_survivors[non_survivors['sex'] == 'female']['age'], kde=False, bins=20, color='pink', label='Female Non-Survivors', stat='density', alpha=0.6, multiple="stack")

# Title and labels
plt.title('Pyramid Histogram of Age and Survival by Sex')
plt.xlabel('Age')
plt.ylabel('Density')
plt.legend(title='Survival Status and Sex')

# Show plot
plt.show()
```

### Question: What can you infer from the age and sex distribution in relation to survival rates?

The pyramid-style histogram reveals clear patterns in survival rates across age and sex groups. The distribution shows that:

1. **Female Survivors vs Non-Survivors**: The number of female survivors is higher in younger age groups, particularly for children (under 10 years old). The survival rate for women seems to be higher across most age groups compared to men, especially in the adult category.
   
2. **Male Survivors vs Non-Survivors**: Male non-survivors show a more pronounced presence in the adult age range, with a lower survival rate overall. Males, particularly in the 20-40 age group, are more likely to be non-survivors compared to females, indicating that women and children were likely prioritized during the evacuation process.

3. **Age Trends**: In both sexes, children had higher survival rates, which might reflect the historical practice of prioritizing children and women in lifeboat access. However, for older adults, the survival rate is lower, which may reflect both physical limitations and the chaotic nature of the disaster.

In summary, the pyramid histogram highlights that women and children had higher survival rates, which was likely influenced by the prioritized lifeboat allocation and rescue efforts, while adult males, especially those in the prime working age range, had lower survival rates.


```python
# Bar Graph - Survival Rates by Passenger Class
plt.figure(figsize=(8, 6))

# Calculate survival rate for each passenger class
survival_rate_by_class = titanic.groupby('pclass')['survived'].mean()

# Create bar plot
sns.barplot(x=survival_rate_by_class.index, y=survival_rate_by_class.values, palette='Blues')

# Title and labels
plt.title('Survival Rates by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Survival Rate')

# Show plot
plt.show()
```

```python
# Line Plot - Average Fare by Passenger Class
plt.figure(figsize=(8, 6))

# Calculate average fare for each passenger class
average_fare_by_class = titanic.groupby('pclass')['fare'].mean()

# Create line plot
sns.lineplot(x=average_fare_by_class.index, y=average_fare_by_class.values, marker='o', color='green')

# Title and labels
plt.title('Average Fare by Passenger Class')
plt.xlabel('Passenger Class')
plt.ylabel('Average Fare')

# Show plot
plt.show()
```

```python
# Scatter Plot - Age vs. Fare with Survival Status
plt.figure(figsize=(10, 6))

# Scatter plot of age vs fare, colored by survival status
sns.scatterplot(x='age', y='fare', data=titanic, hue='survived', palette='coolwarm', alpha=0.6)

# Title and labels
plt.title('Age vs. Fare with Survival Status')
plt.xlabel('Age')
plt.ylabel('Fare')

# Show plot
plt.show()
```

   - The bar graph clearly shows that survival rates were higher for 1st-class passengers and significantly lower for 3rd-class passengers. This reflects the socioeconomic disparities on the Titanic, where wealthier passengers had better access to lifeboats and other means of survival.

2. **Line Plot (Average Fare by Passenger Class)**:
   - The line plot demonstrates that 1st-class passengers paid significantly higher fares compared to 2nd- and 3rd-class passengers. This supports the idea that fare is closely tied to the socioeconomic status of the passengers, with wealthier passengers having better accommodations and, in turn, higher survival rates.

3. **Scatter Plot (Age vs. Fare with Survival Status)**:
   - The scatter plot reveals that younger passengers (especially children) had higher survival rates, regardless of their fare, and this was often more prominent in wealthier passengers (1st class).
   - It also shows that while age and fare have some correlation, survival was likely influenced more by factors like class and gender (with women and children prioritized).

These visualizations offer additional perspectives on the Titanic dataset, highlighting the importance of class and socioeconomic status in survival chances. The bar graph emphasizes class-based survival disparities, the line plot underscores fare differences, and the scatter plot helps further understand how age and fare played a role in survival.
