```python
import pandas as pd

# Use the correct file name
stocks_df = pd.read_csv('Stocks (1).csv')

# Clean column names
stocks_df.columns = stocks_df.columns.str.strip()

# Remove dollar symbols and convert to numeric
stocks_df['Purchased Price'] = stocks_df['Purchased Price'].replace('[\$,]', '', regex=True).astype(float)
stocks_df['Current Price'] = stocks_df['Current Price'].replace('[\$,]', '', regex=True).astype(float)
stocks_df['Shares'] = stocks_df['Shares'].astype(int)

# Convert dates to MM/DD/YY format
stocks_df['Purchased Date'] = pd.to_datetime(stocks_df['Purchased Date']).dt.strftime('%m/%d/%y')

# Show cleaned data
stocks_df.head()

```

```python
# Create portfolio: (Purchased Date, Purchased Price, Shares, Symbol, Current Price)
portfolio = list(stocks_df.itertuples(index=False, name=None))

# Show first 5 entries to verify
portfolio[:5]
```

Functions to Analyze the Portfolio

```python
# Total Cost of Entire Portfolio
def total_portfolio_cost(portfolio):
    return sum(block[1] * block[2] for block in portfolio)

# Total Cost for a Specific Stock
def total_stock_cost(portfolio, ticker):
    return sum(block[1] * block[2] for block in portfolio if block[3] == ticker)

# Overall Profit or Loss Calculation
def overall_profit_loss(portfolio):
    return sum((block[4] - block[1]) * block[2] for block in portfolio)
```

Run the Calculations

```python
# Total Portfolio Cost
total_cost = total_portfolio_cost(portfolio)
print(f"Total Portfolio Cost: ${total_cost:,.2f}")

# Example: Total Cost for GM Stock (replace 'GM' with another symbol if GM not present)
gm_cost = total_stock_cost(portfolio, 'GM')
print(f"Total Cost for GM Stock: ${gm_cost:,.2f}")

# Overall Profit or Loss
profit_loss = overall_profit_loss(portfolio)
print(f"Overall Profit/Loss: ${profit_loss:,.2f}")
```
