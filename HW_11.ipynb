```python
# Author: Farah Makkawi

import matplotlib.pyplot as plt

# Data for the chart
cities = ['Rome', 'Madrid', 'Istanbul', 'Houston']
measurements = [6, 1, 3, 5]

plt.figure(figsize=(8, 5))
bars = plt.bar(cities, measurements, color='cornflowerblue')

# Add title and labels
plt.title("City Measurement Overview")
plt.xlabel("Cities")
plt.ylabel("Measurement")

# Add annotation to the highest bar (Rome)
plt.annotate('highest',
             xy=('Rome', 6),
             xytext=('Rome', 7),
             ha='center',
             arrowprops=dict(facecolor='black', shrink=0.05))

# Add label for daily average
plt.text(3.5, 8.5, "Daily Average", fontsize=9, style='italic', bbox=dict(facecolor='lightgray'))

plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python

import pandas as pd

# Load the data and clean it
df = pd.read_csv("carsData.csv", header=None)
df.columns = ["mpg", "engine", "horse", "weight", "acceleration", "year", "origin", "cylinder"]
df = df[1:].copy()  # Remove row with actual headers inside the data
df = df.apply(pd.to_numeric, errors='coerce')

# Group by year to get average weight
avg_weight_by_year = df.groupby("year")["weight"].mean()

# Plot
plt.figure(figsize=(8, 5))
avg_weight_by_year.plot(kind='line', marker='o', color='green')
plt.title("Average Car Weight by Year")
plt.xlabel("Year")
plt.ylabel("Average Weight (lbs)")
plt.grid(True)
plt.tight_layout()
plt.show()
```
