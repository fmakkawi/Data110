Introduction
This dataset tracks the number of participants and the total costs of the U.S. Food Stamps Program (SNAP) from 1969 to 2015. It includes:

Year: The year of the data.

Participants: Number of individuals in millions.

Costs: Program costs in billions.

The goal of this analysis is to explore the relationship between the number of participants and the costs over time. By analyzing trends and correlations, we aim to understand how the program has grown and its financial impact, providing insights into social and economic factors influencing the need for food assistance.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# URL of the raw dataset on GitHub
url = 'https://raw.githubusercontent.com/fmakkawi/Data110/main/food_stamps.csv'
```

```python
# Load the dataset into a pandas DataFrame
df = pd.read_csv(url)
```

```python
# Display first few rows to inspect data
df.head()
```

```python
# Check for missing values
df.isnull().sum()
```

```python
# Summary statistics
df.describe()
```

```python
# Check for correlations
df.corr()
```

```python
#Line Chart for Trends Over Time

plt.figure(figsize=(10, 6))
plt.plot(df['year'], df['participants'], label='Participants (millions)', color='blue')
plt.plot(df['year'], df['costs'], label='Costs (billions)', color='red')
plt.xlabel('Year')
plt.ylabel('Value')
plt.title('Food Stamp Participation and Costs Over Time')
plt.legend()
plt.show()
```

```python
#Scatter Plot for Relationship Between Participants and Costs
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='participants', y='costs')
plt.title('Relationship Between Participants and Costs')
plt.xlabel('Participants (millions)')
plt.ylabel('Costs (billions)')
plt.show()

```

```python
#Bar Chart for Decadal Cost Comparison
# Extract decade
df['decade'] = (df['year'] // 10) * 10

# Group by decade and sum costs
decade_costs = df.groupby('decade')['costs'].sum().reset_index()

plt.figure(figsize=(10, 6))
sns.barplot(x='decade', y='costs', data=decade_costs)
plt.title('Decadal Comparison of Costs')
plt.xlabel('Decade')
plt.ylabel('Total Costs (billions)')
plt.show()
```

# Conclusion:
The analysis shows a clear upward trend in both food stamp participation and costs over time. Notably, the Great Recession in 2008 correlates with a significant spike in both metrics. The relationship between participation and costs is strongly positive, suggesting that as more individuals rely on food assistance, the costs to the government increase.
